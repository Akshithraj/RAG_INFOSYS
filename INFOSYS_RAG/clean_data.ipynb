```python
# =========================
# PROFIT LOSS
# =========================

profit_df = spark.table(
    "enterprise_ai.raw_data.profit_loss"
)

profit_df = profit_df \
    .withColumnRenamed(" FY 26 ", "FY26") \
    .withColumnRenamed(" FY 25 ", "FY25") \
    .withColumnRenamed(" FY 24 ", "FY24")

profit_df.write \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable(
        "enterprise_ai.raw_data.profit_loss_clean"
    )

print("profit_loss cleaned")


# =========================
# BALANCE SHEET
# =========================

balance_df = spark.table(
    "enterprise_ai.raw_data.balance_sheet"
)

balance_df = balance_df \
    .withColumnRenamed(" FY 26 ", "FY26") \
    .withColumnRenamed(" FY 25 ", "FY25") \
    .withColumnRenamed(" FY 24 ", "FY24")

balance_df.write \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable(
        "enterprise_ai.raw_data.balance_sheet_clean"
    )

print("balance_sheet cleaned")


# =========================
# EMPLOYEE DATA
# =========================

employee_df = spark.table(
    "enterprise_ai.raw_data.employee_data"
)

employee_df = employee_df \
    .withColumnRenamed(" FY 26 ", "FY26") \
    .withColumnRenamed(" FY 25 ", "FY25") \
    .withColumnRenamed(" FY 24 ", "FY24")

employee_df.write \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable(
        "enterprise_ai.raw_data.employee_data_clean"
    )

print("employee_data cleaned")
```
