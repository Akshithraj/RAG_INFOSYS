```python
# =========================
# FINANCIAL TOOL
# =========================

spark.sql("""

CREATE OR REPLACE FUNCTION enterprise_ai.raw_data.get_financial_value(

    metric_name STRING,
    year_name STRING

)

RETURNS STRING

LANGUAGE SQL

RETURN (

    CASE

        WHEN year_name = 'FY26' THEN (

            SELECT CAST(FY26 AS STRING)

            FROM enterprise_ai.raw_data.profit_loss_clean

            WHERE lower(Particulars) = lower(metric_name)

            LIMIT 1
        )

        WHEN year_name = 'FY25' THEN (

            SELECT CAST(FY25 AS STRING)

            FROM enterprise_ai.raw_data.profit_loss_clean

            WHERE lower(Particulars) = lower(metric_name)

            LIMIT 1
        )

        WHEN year_name = 'FY24' THEN (

            SELECT CAST(FY24 AS STRING)

            FROM enterprise_ai.raw_data.profit_loss_clean

            WHERE lower(Particulars) = lower(metric_name)

            LIMIT 1
        )

    END

);

""")

print("Financial tool created")
```

```python
spark.sql("""

CREATE OR REPLACE FUNCTION enterprise_ai.raw_data.search_company_documents(

    search_text STRING COMMENT 'Search keyword like AI, cloud, revenue, hiring'

)

RETURNS TABLE (

    company STRING,
    financial_year STRING,
    matched_text STRING

)

COMMENT 'Searches Infosys annual reports'

LANGUAGE SQL

RETURN (

    SELECT
        company,
        financial_year,

        substring(
            product_doc,
            greatest(
                locate(lower(search_text), lower(product_doc)) - 300,
                1
            ),
            1000
        ) as matched_text

    FROM enterprise_ai.raw_data.product_docs

    WHERE lower(product_doc)
    LIKE concat('%', lower(search_text), '%')

    LIMIT 3

);

""")
```

```python
spark.sql("""

CREATE OR REPLACE FUNCTION enterprise_ai.raw_data.get_company_metric(

    input_metric STRING,
    year_name STRING

)

RETURNS STRING

LANGUAGE SQL

RETURN (

    CASE

        WHEN year_name = 'FY26' THEN (

            SELECT COALESCE(
                CAST(FY26 AS STRING),
                'Metric not found'
            )

            FROM enterprise_ai.raw_data.company_master

            WHERE lower(metric_name)
            = lower(input_metric)

            LIMIT 1
        )

        WHEN year_name = 'FY25' THEN (

            SELECT COALESCE(
                CAST(FY25 AS STRING),
                'Metric not found'
            )

            FROM enterprise_ai.raw_data.company_master

            WHERE lower(metric_name)
            = lower(input_metric)

            LIMIT 1
        )

        WHEN year_name = 'FY24' THEN (

            SELECT COALESCE(
                CAST(FY24 AS STRING),
                'Metric not found'
            )

            FROM enterprise_ai.raw_data.company_master

            WHERE lower(metric_name)
            = lower(input_metric)

            LIMIT 1
        )

    END

);

""")

print("Universal tool recreated")
```
